```python
import pickle
import pandas as pd
import feather
```

```python
with open("../data/input/data.covs.pickle", "rb") as f:
    df = pickle.load(f)
```

```python
df.head()
```

```python
df.shape
```

```python
df["molecule_name"].nunique()
```

```python
cols = df.columns.values[6:]
```

```python
cols
```

```python
train = feather.read_dataframe("../data/input/train.feather")
test = feather.read_dataframe("../data/input/test.feather")
```

```python
train.head()
```

```python
train = train.merge(df, on=["molecule_name", "atom_index_0", "atom_index_1"], how="left")
test = test.merge(df, on=["molecule_name", "atom_index_0", "atom_index_1"], how="left")
```

```python
train.head()
```

```python
qm_train = train[cols]
qm_test = test[cols]
```

```python
qm_train.shape, qm_test.shape
```

```python
qm_train.to_feather('../features/QM_train.feather')
qm_test.to_feather('../features/QM_test.feather')
```
